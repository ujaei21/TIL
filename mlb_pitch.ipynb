```python
ryu_stats = statcast_pitcher('2021-04-01', '2021-10-3',547943)
ryu_stats.head(5)
```

```python
ryu_stats.game_date.nunique()
```

```python
result=ryu_stats[['pitch_name','game_date','release_speed','release_pos_x','release_pos_y','release_pos_z','release_spin_rate','release_extension','spin_axis']]
```

```python
import pandas as pd
```

```python
pd.DataFrame(result)
```

```python
result.pitch_name.unique()
```

```python
result.groupby(['pitch_name','game_date']).std().round(2)
```
