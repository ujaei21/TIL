```python
import os
from selenium import webdriver 
import time 
import pandas as pd 

webdriver_path = os.getcwd()+'.//webdriver//chromedriver.exe'

url_base = 'https://www.koreabaseball.com/Record/Player/' 
category_list = ['HitterBasic/Basic1.aspx', 'Defense/Basic.aspx'] 
driver = webdriver.Chrome(webdriver_path) 
driver.implicitly_wait(3) 
# driver.maximize_window() 

# 드라이버 열기
driver.get(url_base+category_list[0])

# 정규시즌으로 페이지 넘김
driver.find_element_by_class_name('select02').click()
xpath_format = '//option[@value="0"]'
driver.find_element_by_xpath(xpath_format).click()
time.sleep(0.5)

# Dataframe으로 변환
temp_list = [x.split(' ') for x in driver.find_element_by_tag_name('table').text.split('\n')] 
temp_data = pd.DataFrame(columns = temp_list[0]) 
for i,temp in enumerate(temp_list[1:]): 
    # 투수 기록 대비
    try: 
        temp_data.loc[i] = temp 
        time.sleep(0.5) 
    except: 
        ip_index = temp_list[0].index('IP') 
        ip = temp[ip_index] 
        del temp[13] 
        temp[13] = ip+' '+temp[13] 
        temp_data.loc[i] = temp 
        time.sleep(0.5)

baseball_data = temp_data.copy()
# 넥스트 버튼 누르기
driver.find_element_by_class_name('next').click()
# Dataframe으로 변환
temp_list = [x.split(' ') for x in driver.find_element_by_tag_name('table').text.split('\n')] 
temp_data = pd.DataFrame(columns = temp_list[0]) 
for i,temp in enumerate(temp_list[1:]): 
    # 투수 기록 대비
    try: 
        temp_data.loc[i] = temp 
        time.sleep(0.5) 
    except: 
        ip_index = temp_list[0].index('IP') 
        ip = temp[ip_index] 
        del temp[13] 
        temp[13] = ip+' '+temp[13] 
        temp_data.loc[i] = temp 
        time.sleep(0.5)
    

baseball_data_test = pd.merge(baseball_data,temp_data,on = ['순위','선수명','팀명','AVG'])
# x_path 만들기
xpath_format = '//*[@id="cphContents_cphContents_cphContents_udpContent"]/div[2]/div[1]/ul/li[2]'
driver.find_element_by_xpath(xpath_format).click()

# Dataframe으로 변환
temp_list = [x.split(' ') for x in driver.find_element_by_tag_name('table').text.split('\n')] 
temp_data = pd.DataFrame(columns = 
                         ['순위','선수명','팀명','AVG','XBH','GO','AO','GO/AO','GW RBI','BB/K','P/PA','ISOP','XR','GPA']) 
for i,temp in enumerate(temp_list[1:]): 
    # 투수 기록 대비
    try: 
        temp_data.loc[i] = temp 
        time.sleep(0.5) 
    except: 
        ip_index = temp_list[0].index('IP') 
        ip = temp[ip_index] 
        del temp[13] 
        temp[13] = ip+' '+temp[13] 
        temp_data.loc[i] = temp 
        time.sleep(0.5)

baseball_data = pd.merge(baseball_data_test,temp_data,on = ['순위','선수명','팀명','AVG'])
```
