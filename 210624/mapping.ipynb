```python
import pandas as pd
import folium as g
```

```python
data=pd.read_csv('Measurement_summary.csv')
```

```python
data.columns
```

```python
data.drop(['Measurement date','Address','SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5'],axis=1)
```

```python
df_loc=pd.DataFrame({'장소':data["Station code"].unique(),
                     '위도':data.Latitude.unique(),'경도':data.Longitude.unique()})
```

```python
df_loc.head()
```

```python
lat=df_loc.iloc[:,1].mean()
long=df_loc.iloc[:,2].mean()
```

```python
m=g.Map([lat,long],zoom_start=10)
m
```

```python
for i in range(df_loc.shape[0]):
    sub_lat  = df_loc.iloc[i,1]
    sub_long = df_loc.iloc[i,2]
    
    title=df_loc.iloc[i,0]
    g.Marker([sub_lat,sub_long],tooltip=title,).add_to(m)
```

```python
m
```

```python
m=g.Map([lat,long],zoom_start=10)
```

```python
for i in range(df_loc.shape[0]):
    sub_lat  = df_loc.iloc[i,1]
    sub_long = df_loc.iloc[i,2]
    
    title=df_loc.iloc[i,0]
    g.CircleMarker([sub_lat,sub_long],
                   radius=10,
                   color='red',
                   tooltip=title,).add_to(m)
```

```python
m
```

```python
m=g.Map([lat,long],zoom_start=10)
for i in range(df_loc.shape[0]):
    sub_lat  = df_loc.iloc[i,1]
    sub_long = df_loc.iloc[i,2]
    
    title=df_loc.iloc[i,0]
    g.Marker([sub_lat,sub_long]
             ,icon=g.Icon(color='green')
             ,tooltip=title,).add_to(m)
```

```python
m
```

```python
m.save('center_location.html')
```
