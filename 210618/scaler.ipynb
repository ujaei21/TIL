## RFE

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data=df.values
```

```python
sc=StandardScaler()
```

```python
data_sc=sc.fit_transform(data)
```

```python
data_sc
```

```python
df_sc = pd.DataFrame(data_sc, columns = ['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size',
       'tip_rate'] )
```

```python
df_sc.describe()
```

```python
y=df_sc['tip']
```

```python
df_sc.drop('tip',axis=1,inplace=True)
```

```python
data_sc=df_sc.values
```

```python
from sklearn.svm import SVR
```

```python
estimator = SVR(kernel="linear")
```

```python
rfe = RFE(estimator, n_features_to_select=4)
```

```python
rfe.fit(df_sc,y)
```

```python
for i in range(data_sc.shape[1]):
  print('Column: %10s, Rank: %d' % (df_sc.columns[i], rfe.ranking_[i]))
```

```python
df_rfe=pd.DataFrame(df_sc,columns=df_sc.columns)
```

```python
df_rfe
```

```python
df_rfe.drop(['sex', 'smoker', 'day' ], axis=1, inplace=True)
```
