```python
import pandas as pd
import folium as g
```

```python
data=pd.read_excel('C:/Users/SEC/OneDrive/바탕 화면/station.xlsx')
```

```python
data.head()
```

```python
lat=data.iloc[:,1].mean()
lon=data.iloc[:,2].mean()

m=g.Map([lat,lon],zoom_start=10)
```

```python
for i in range(data.shape[0]):
    sub_lat  = data.iloc[i,1]
    sub_long = data.iloc[i,2]
    
    title=data.iloc[i,0]
    g.Marker([sub_lat,sub_long],tooltip=title,).add_to(m)
```
