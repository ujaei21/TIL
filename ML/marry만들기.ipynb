```python
import pandas as pd
url = '/content/drive/MyDrive/Colab Notebooks/apart/report.txt'

marry = pd.read_csv(url,error_bad_lines=False,sep='\t')
marry.head()
```

```python
marry.drop('이혼',axis=1,inplace=True)
```

```python
marry=marry[marry['자치구']!='합계']
```

```python
marry=marry[marry['월별']!='계']
```

```python
marry['월별']=marry['월별'].str[:-1]
```

```python
marry=marry[['기간','월별','자치구','혼인']]
```

```python
marry=marry.rename({'기간':'year','월별':'month','자치구':'Gu','혼인':'marriage'},axis='columns')
```

```python
marry
```

```python
marry.month=marry.month.astype(int)
marry.marriage=marry.marriage.astype(int)
```

```python
marry.info()
```

```python
marriage_avg=[]
for avg in range(len(marry.month)):
    if marry.month.iloc[avg] in [1,3,5,7,8,10,12]:
        marriage_avg = marry.marriage/31
    elif marry.month.iloc[avg] in [4,6,9,11]:
        marriage_avg = marry.marriage/30
    else:
        marriage_avg = marry.marriage/28
```

```python
marry.month.iloc[0]
```

```python
marriage_avg
```

```python
marry['avg']=marriage_avg
```
