# Convolution Neural Network
합성곱 신경망
- 이미지 처리 작업에 주로 사용

 가중치의 수를 줄이고 연산량을 감소한다. 그럼에도 성능은 좋다.

- 여러개의 필터를 사용해서 이미지의 특징을 추출
- 2차원의 형태로 input을 받을 수 있다.

## CNN hyperparameter

- Filter: 필터와 이미지의 행과 열의 위치가 같은 숫자를 곱해서 더한다. 
- Stride: 필터를 적용하기 위해 이동하는 위치의 간격, 값이 커지면 맵이 축소된다. 많이 사용되진 않는다.
-Pooling: 풀링 사이즈를 정해서 특징맵에서 숫자 하나만 뽑는다.
-Padding: 출력의 크기를 조정할 목적으로 사용, input_data 주변을 0으로 채우는 것




```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.datasets import mnist
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

### 1)reshape Nomalization

```python
X_train = X_train.reshape((60000,28,28,1))
X_test = X_test.reshape((10000,28,28,1))

X_train.shape, X_test.shape
```

```python
X_train = X_train.astype(float)/255
X_test = X_test.astype(float)/255
```

### 2) One Hot Encoding

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

## 2.Keras Modeling

```python
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential()
model.add(layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)))
model.add(layers.MaxPool2D(pool_size=(2,2)))
model.add(layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
model.add(layers.MaxPool2D(pool_size=(2,2)))
model.add(layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
```

```python
model.summary()
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(units=64,activation='relu'))
model.add(layers.Dense(units=10,activation='softmax'))
```

```python
model.summary()
```

### 2) Compile

```python
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

### 3) fit

```python
%%time

Hist_mnist = model.fit(X_train, y_train,epochs = 100,batch_size = 128,validation_split = 0.2)
```

```python
import matplotlib.pyplot as plt

epochs = range(1,len(Hist_mnist.history['loss'])+1)

plt.figure(figsize=(9,6))
plt.plot(epochs, Hist_mnist.history['loss'])
plt.plot(epochs, Hist_mnist.history['val_loss'])
# plt.ylim(0, 0.4)
plt.title('Training & Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Training Loss', 'Validation Loss'])
plt.grid()
plt.show()
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print('Loss = {:.5f}'.format(loss))
print('Accuracy = {:.5f}'.format(accuracy))
```
